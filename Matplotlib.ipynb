**(d) Plot subplot using scatter plot**

```python
# Create a grid of four subplots within a single figure
fig, axs = plt.subplots(nrows=2, ncols=2,figsize=(12,9))

# Plot the first graph in the upper left subplot
axs[0,0].scatter(data["Model"],data["cyl"],color='green')
axs[0,0].set_title('car Model vs Cylinder',color='Red')
axs[0,0].set_xticks(data["Model"])
axs[0,0].set_xticklabels(data["Model"],rotation=90)
axs[0,0].set_xlabel("Model",color="blue")
axs[0,0].set_ylabel("Cylinder",color="blue")

# Plot the second graph in the upper right subplot
axs[0,1].scatter(data["Model"],data["mpg"],color='yellow')
axs[0,1].set_title('car Model vs Miles per Galon',color='Red')
axs[0,1].set_xticks(data["Model"])
axs[0,1].set_xticklabels(data["Model"],rotation=90)
axs[0,1].set_xlabel("Model",color="blue")
axs[0,1].set_ylabel("Miles per Galon",color="blue")

# Plot the Third graph in the bottom left subplot
axs[1,0].scatter(data["Model"],data["mpg"],color='violet')
axs[1,0].set_title('car Model vs rear axle ratio',color='Red')
axs[1,0].set_xticks(data["Model"])
axs[1,0].set_xticklabels(data["Model"],rotation=90)
axs[1,0].set_xlabel("Model",color="blue")
axs[1,0].set_ylabel("Rear axle ratio",color="blue")

# Plot the fourth graph in the bottom right subplot
axs[1,1].scatter(data["Model"],data["wt"],color='orange')
axs[1,1].set_title('car Model vs Vehicle Weight',color='Red')
axs[1,1].set_xticks(data["Model"])
axs[1,1].set_xticklabels(data["Model"],rotation=90)
axs[1,1].set_xlabel("Model",color="blue")
axs[1,1].set_ylabel("Weight",color="blue")

fig.suptitle('Car Model and its Specifications',color='black',size=20)
plt.tight_layout()
plt.show()
```
