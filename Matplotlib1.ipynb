**Creating different bar charts in one image using sub-plot**

```python
ecom=["Myntra","Snapdeal","Alibabas","Amazon","Flipkart"]
q1_profit=[35,45,100,75,40]
q2_profit=[83,83,37,28,78]
q3_profit=[67,73,93,60,61]
q4_profit=[20,85,92,90,22]

plt.figure(1,figsize=(10,10)) ##sets size of plot ,default size is (4,4)

#creating bar chart on first cell of figure with 2 rows and 2 columns
plt.subplot(221)                                                        #221 represents, 2 rows 2 columns in 1 st block of image
plt.bar(ecom,q1_profit)
plt.title("Quarter 1 profit")

plt.subplot(222)
plt.bar(ecom,q2_profit,color="red")
plt.title("Quarter 2 profit")

plt.subplot(223)
plt.bar(ecom,q3_profit,color="orange")
plt.title("Quarter 3 profit")

plt.subplot(224)
plt.bar(ecom,q4_profit,color="yellow")
plt.title("Quarter 4 profit")
```

**Program to create stack bar shart**

```python
plt.figure(1,figsize=(15,10))
Country=["Las vegas","paris","London","Austria","Norway"]
pop_1930=[5,2,18,5,10]
pop_1940=[10,5,18,13,20]
pop_1950=[15,12,20,25,30]
pop_1960=[20,15,32,33,50]
pop_1970=[25,18,42,45,60]
pop_1980=[30,25,45,50,70]
pop_1990=[35,35,49,55,80]
pop_2000=[40,50,55,60,90]

plt.bar(Country,pop_2000,color="pink")
plt.bar(Country,pop_1990,color="red")
plt.bar(Country,pop_1980,color="orange")
plt.bar(Country,pop_1970,color="yellow")
plt.bar(Country,pop_1960,color="green")
plt.bar(Country,pop_1950,color="blue")
plt.bar(Country,pop_1940,color="indigo")
plt.bar(Country,pop_1930,color="violet")
```
